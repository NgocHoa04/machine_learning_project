## Train-Test Split

```python
# Chia train-test split theo th·ªùi gian, trong ƒë√≥ kho·∫£ng c√°ch t·∫≠p train v√† test l√† 7 ng√†y ƒë·ªÉ tr√°nh data leakage do t√≠nh ch·∫•t th·ªùi gian c·ªßa d·ªØ li·ªáu, temp l√† bi·∫øn m·ª•c ti√™u
def tran_val_test_split(df, train_ratio):
    train_size = int(len(df) * train_ratio)
    train_data = df.iloc[:train_size] 

    test_data = df.iloc[train_size:]
    print(f"K√≠ch th∆∞·ªõc t·∫≠p hu·∫•n luy·ªán: {train_data.shape}")
    print(f"K√≠ch th∆∞·ªõc t·∫≠p ki·ªÉm tra: {test_data.shape}")
    print(f"Kho·∫£ng th·ªùi gian t·∫≠p hu·∫•n luy·ªán: {train_data['datetime'].min()} ƒë·∫øn {train_data['datetime'].max()}")
    print(f"Kho·∫£ng th·ªùi gian t·∫≠p ki·ªÉm tra: {test_data['datetime'].min()} ƒë·∫øn {test_data['datetime'].max()}")

    # temp l√† bi·∫øn m·ª•c ti√™u
    feature_cols = df.columns.drop(['temp', 'datetime'])  # Lo·∫°i b·ªè c·ªôt temp kh·ªèi features
    X_train = train_data[feature_cols]
    y_train = train_data['temp']
    X_test = test_data[feature_cols]
    y_test = test_data['temp']

    print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
    print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
    return X_train, y_train, X_test, y_test
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.feature_selection import VarianceThreshold
from sklearn.pipeline import Pipeline
import numpy as np

# B∆∞·ªõc 1: Lo·∫°i b·ªè c·ªôt s·ªë c√≥ ph∆∞∆°ng sai = 0
class VarianceThresholdSelector(BaseEstimator, TransformerMixin):
    def __init__(self, threshold=0):
        self.threshold = threshold
        self.selector = VarianceThreshold(threshold=self.threshold)
    
    def fit(self, X, y=None):
        numeric_cols = X.select_dtypes(include=['float64', 'int64']).columns
        self.numeric_cols = numeric_cols
        self.selector.fit(X[numeric_cols])
        # L∆∞u l·∫°i c·ªôt s·ªë ƒë∆∞·ª£c gi·ªØ l·∫°i
        self.retained_cols = numeric_cols[self.selector.get_support()]
        # L∆∞u l·∫°i c·ªôt s·ªë b·ªã lo·∫°i
        self.dropped_cols = [col for col in numeric_cols if col not in self.retained_cols]
        print("C√°c c·ªôt b·ªã lo·∫°i v√¨ ph∆∞∆°ng sai = 0:", self.dropped_cols)
        return self
    
    def transform(self, X):
        cols_to_keep = list(self.retained_cols) + list(X.columns.difference(self.numeric_cols))
        return X[cols_to_keep]

# B∆∞·ªõc 2: Lo·∫°i b·ªè c·ªôt h·∫±ng s·ªë v√† d√≤ng tr√πng l·∫∑p
class ConstantAndDuplicateRemover(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        # X√°c ƒë·ªãnh c·ªôt h·∫±ng s·ªë
        self.constant_cols = [col for col in X.columns if X[col].nunique() == 1]
        print("C√°c c·ªôt b·ªã lo·∫°i v√¨ ch·ªâ c√≥ 1 gi√° tr·ªã:", self.constant_cols)
        return self
    
    def transform(self, X):
        # B·ªè c·ªôt h·∫±ng s·ªë
        X_cleaned = X.drop(columns=self.constant_cols, errors="ignore")
        # B·ªè d√≤ng tr√πng l·∫∑p (kh√¥ng in s·ªë l∆∞·ª£ng n·ªØa)
        X_cleaned = X_cleaned.drop_duplicates()
        return X_cleaned

    
# T·∫°o pipeline
cleaning_pipeline = Pipeline(steps=[
    ("var_thresh", VarianceThresholdSelector(threshold=0)),
    ("const_dup", ConstantAndDuplicateRemover())

])
X_cleaned = cleaning_pipeline.fit_transform(X_train.copy())
print(f"K√≠ch th∆∞·ªõc d·ªØ li·ªáu sau khi l√†m s·∫°ch: {X_cleaned.shape}")
```

## Correlation Analysis for Categorical and Mixed Variables

### 1. Categorical-Categorical Correlation: Cram√©r's V
Cram√©r's V ƒëo l∆∞·ªùng m·ªëi quan h·ªá gi·ªØa 2 bi·∫øn categorical d·ª±a tr√™n chi-square test. Gi√° tr·ªã t·ª´ 0 (kh√¥ng c√≥ m·ªëi quan h·ªá) ƒë·∫øn 1 (m·ªëi quan h·ªá ho√†n h·∫£o).

### 2. Numerical-Categorical Correlation: Correlation Ratio (Œ∑¬≤)
Correlation ratio (eta-squared) ƒëo l∆∞·ªùng m·ª©c ƒë·ªô m·ªôt bi·∫øn categorical gi·∫£i th√≠ch ƒë∆∞·ª£c s·ª± bi·∫øn thi√™n c·ªßa m·ªôt bi·∫øn numerical. Gi√° tr·ªã t·ª´ 0 ƒë·∫øn 1.

```python
# Import th∆∞ vi·ªán c·∫ßn thi·∫øt
import numpy as np
from scipy.stats import chi2_contingency
import warnings
warnings.filterwarnings('ignore')

# H√†m t√≠nh Cram√©r's V cho categorical-categorical correlation
def cramers_v(x, y):
    """
    T√≠nh Cram√©r's V gi·ªØa 2 bi·∫øn categorical
    V = sqrt(chi2 / (n * (min(k,r) - 1)))
    """
    # Lo·∫°i b·ªè missing values
    mask = ~(pd.isna(x) | pd.isna(y))
    x_clean = x[mask]
    y_clean = y[mask]
    
    if len(x_clean) == 0:
        return np.nan
    
    # T·∫°o contingency table
    confusion_matrix = pd.crosstab(x_clean, y_clean)
    
    # Chi-square test
    chi2, _, _, _ = chi2_contingency(confusion_matrix)
    
    n = confusion_matrix.sum().sum()
    min_dim = min(confusion_matrix.shape[0], confusion_matrix.shape[1]) - 1
    
    if min_dim == 0:
        return 0
    
    v = np.sqrt(chi2 / (n * min_dim))
    return v

# H√†m t√≠nh correlation ratio (eta-squared) cho numerical-categorical correlation
def correlation_ratio(categories, values):
    """
    T√≠nh correlation ratio (Œ∑¬≤) gi·ªØa bi·∫øn categorical v√† numerical
    Œ∑¬≤ = SSB / SST (Sum of Squares Between / Sum of Squares Total)
    """
    # Lo·∫°i b·ªè missing values
    mask = ~(pd.isna(categories) | pd.isna(values))
    categories_clean = categories[mask]
    values_clean = values[mask]
    
    if len(values_clean) == 0:
        return np.nan
    
    # T√≠nh mean t·ªïng th·ªÉ
    mean_total = values_clean.mean()
    
    # T√≠nh SST (Total Sum of Squares)
    sst = np.sum((values_clean - mean_total) ** 2)
    
    if sst == 0:
        return 0
    
    # T√≠nh SSB (Between group Sum of Squares)
    ssb = 0
    for category in categories_clean.unique():
        group_values = values_clean[categories_clean == category]
        group_mean = group_values.mean()
        group_size = len(group_values)
        ssb += group_size * (group_mean - mean_total) ** 2
    
    # Œ∑¬≤ = SSB / SST
    eta_squared = ssb / sst
    return eta_squared

print("‚úÖ ƒê√£ ƒë·ªãnh nghƒ©a c√°c h√†m t√≠nh correlation")
```

### T√≠nh Categorical-Categorical Correlation (Cram√©r's V)

```python
# C·∫≠p nh·∫≠t l·∫°i categorical_cols ƒë·ªÉ ch·ªâ gi·ªØ c√°c c·ªôt c√≤n t·ªìn t·∫°i trong df
categorical_cols_cleaned = [col for col in categorical_cols if col in X_cleaned.columns]

# T√≠nh Cram√©r's V matrix cho t·∫•t c·∫£ c√°c c·∫∑p bi·∫øn categorical c√≤n l·∫°i
if len(categorical_cols_cleaned) > 0:
    cat_cat_corr = pd.DataFrame(
        np.zeros((len(categorical_cols_cleaned), len(categorical_cols_cleaned))),
        index=categorical_cols_cleaned,
        columns=categorical_cols_cleaned
    )

    for i, col1 in enumerate(categorical_cols_cleaned):
        for j, col2 in enumerate(categorical_cols_cleaned):
            if i <= j:  # Ch·ªâ t√≠nh n·ª≠a tr√™n c·ªßa ma tr·∫≠n (symmetric)
                v = cramers_v(X_cleaned[col1], X_cleaned[col2])
                cat_cat_corr.loc[col1, col2] = v
                cat_cat_corr.loc[col2, col1] = v

    print("üìä Cram√©r's V Matrix (Categorical-Categorical):")
    display(cat_cat_corr)
else:
    print("‚ö†Ô∏è Kh√¥ng c√≤n c·ªôt categorical n√†o trong X_cleaned ƒë·ªÉ ph√¢n t√≠ch!")
```

### üìä Cram√©r's V Analysis: Categorical Variable Associations

**Overview**

Cram√©r's V measures the strength of association between categorical variables using chi-square statistics, with values from 0 (no relationship) to 1 (perfect relationship). This analysis examines six weather features: sunrise, sunset, conditions, description, icon, and stations.

---

**Key Findings**

**1. Perfect Correlations (V = 1.000): Problems to Fix**

**Sunrise ‚Üî Sunset**
- Both variables are calculated from the same inputs: Hanoi's latitude (21¬∞N) and day of year
- They always change together, creating redundancy
- **Action:** Keep only one variable or create a new feature called "day length" (sunset - sunrise)

**Sunrise/Sunset ‚Üî Weather Conditions**
- This correlation is misleading - sunrise time doesn't actually cause weather
- The real connection is through date: each date has unique sunrise time AND weather observation
- This creates a "memorization" problem: the model can't predict future dates with new sunrise times
- **Action:** Remove sunrise/sunset as categorical features; convert to numerical features instead

**Weather Descriptors (conditions, description, icon)**
- V ‚âà 0.999-1.000 shows these three variables contain identical information in different formats
  - conditions: "Rain, Partially cloudy"
  - description: "Partly cloudy throughout the day with rain"
  - icon: "rain"
- All three come from the same weather classification system
- **Action:** Keep only "conditions" variable; drop description and icon

**Stations ‚Üî Sunrise/Sunset**
- V = 1.000 is another false correlation through date
- Stations don't affect sunrise time; date drives both
- **Action:** Remove stations from predictive features

---

**2. Weak Associations (V < 0.1): Safe to Remove**

**Stations ‚Üî Weather Variables**
- V = 0.049-0.081 shows almost no relationship
- **Why:** Weather systems in Hanoi are much larger (100-1000 km) than the city size (58 km diameter)
- All weather stations measure essentially the same conditions
- **Action:** Remove stations - they provide no predictive value for Hanoi's compact geography

---

**Feature Engineering Strategy**

**Transform Sunrise/Sunset**
Instead of categorical strings, create numerical features:
1. **Day length** (hours): Simple and interpretable, ranges 11.5-13.5 hours across seasons
2. **Seasonal encoding**: Use sine/cosine of day-of-year to capture cyclical patterns
3. **Solar hour**: Convert time strings to numerical hours (e.g., 6:45 AM ‚Üí 6.75)

**Simplify Weather Variables**
- Keep: `conditions` (categorical)
- Remove: `description`, `icon` (redundant)
- Alternative: Use only numerical weather variables (precipprob, cloudcover, etc.) and drop all categorical weather descriptors

**Remove Non-Predictive Features**
- Drop: `stations` (no geographic variation in Hanoi)
- Drop: Raw sunrise/sunset strings (spurious correlations)

---
By removing redundant variables and transforming categorical time features to numerical formats, we can build a cleaner, more robust prediction model.

### üìä Correlation Ratio (Œ∑¬≤) Analysis: Numerical-Categorical Relationships

**Overview**

Correlation ratio (Œ∑¬≤) measures how much a categorical variable explains the variation in a numerical variable. Values range from 0 (no explanatory power) to 1 (perfect grouping). This analysis examines relationships between weather categories (sunrise, sunset, conditions) and numerical weather measurements.

---

**Key Findings from Heatmap**

**1. Perfect Correlations (Œ∑¬≤ = 1.000): Spurious Relationships**

**Sunrise/Sunset ‚Üî All Numerical Variables**
- Both show Œ∑¬≤ = 1.000 with every numerical variable (temp, humidity, precipitation, wind, pressure, solar variables)
- **This is NOT a real relationship** - it's a data structure artifact
- **Why it happens:** Each day has a unique sunrise/sunset time string, creating perfect 1-to-1 mapping with that day's weather measurements
- **Problem:** Model will memorize specific dates instead of learning weather patterns
- **Action:** Remove sunrise/sunset categorical variables; convert to numerical features (day length, seasonal cycles)

---

**2. Meaningful Relationships with Weather Conditions**

**Strong Associations (Œ∑¬≤ ‚â• 0.5)**

**Precipitation Probability (Œ∑¬≤ = 1.000)**
- Weather conditions perfectly explain precipitation probability
- Clear days ‚Üí low precipprob (0-10%), Rainy days ‚Üí high precipprob (80-100%)
- This makes logical sense: weather condition categories are partially defined by precipitation
- **Use:** Conditions is an excellent predictor for precipitation-related variables

**Cloud Cover (Œ∑¬≤ = 0.570)**
- Weather conditions explain 57% of cloud cover variation
- Clear ‚Üí 0-20% clouds, Partially Cloudy ‚Üí 20-80% clouds, Overcast/Rain ‚Üí 80-100% clouds
- **Note:** 43% unexplained variance due to overlap within categories (e.g., "Partially Cloudy" has wide range)

---

**Moderate Associations (0.2 < Œ∑¬≤ < 0.5)**

**Humidity (Œ∑¬≤ = 0.310)**
- Weather conditions explain 31% of humidity variation
- **Why relatively low:** Hanoi's subtropical monsoon climate maintains persistently high humidity (75-85%) year-round
- Even "Clear" days often have 70%+ humidity due to geographic location (Red River proximity, monsoon influence)
- Small differences: Clear ‚âà 70-75%, Rain ‚âà 80-90%

**Precipitation Coverage (Œ∑¬≤ = 0.254)**
- Weather categories moderately explain how much area receives rain (25% of variance)
- "Rain" category includes both scattered showers (30% coverage) and widespread rain (90% coverage)
- Considerable overlap reduces predictive power

**Solar Variables (Œ∑¬≤ = 0.231-0.236)**
- Solar radiation, solar energy, and UV index show weak-to-moderate associations with conditions
- **Why:** Cloud cover affects solar radiation, but season and latitude are more important factors
- Winter clear day has less solar radiation than summer cloudy day

---

**Weak Associations (Œ∑¬≤ < 0.2): Limited Predictive Value**

**Temperature Variables (Œ∑¬≤ = 0.14-0.16)**
- Weather conditions poorly explain temperature variation
- **Why:** Temperature is primarily seasonal, not weather-dependent
  - "Rain" in summer: 28-32¬∞C
  - "Rain" in winter: 12-18¬∞C
  - Same category, completely different temperatures
- **Solution:** Need interaction features: `conditions √ó month` or `conditions √ó season`

**Dew Point (Œ∑¬≤ = 0.202)**
- Similar pattern to humidity - relatively constant regardless of weather type
- Hanoi's climate maintains high moisture content year-round

**Sea Level Pressure (Œ∑¬≤ = 0.105)**
- Very weak relationship with weather conditions
- Pressure systems operate on larger spatial scales (synoptic scale: 1000+ km)
- Local weather conditions don't strongly indicate regional pressure patterns

---

**Negligible Associations (Œ∑¬≤ < 0.1): No Predictive Value**

**Wind Variables (Œ∑¬≤ = 0.011-0.025)**
- Weather conditions explain virtually nothing about wind
  - Wind speed: Œ∑¬≤ = 0.025 (2.5% of variance)
  - Wind direction: Œ∑¬≤ = 0.011 (1.1% of variance)
  - Wind gust: Œ∑¬≤ = 0.022 (2.2% of variance)
- **Why:** Hanoi has generally low wind speeds (2-4 m/s average) driven by regional monsoon circulation and pressure gradients, not local weather
- Wind patterns are more related to season (winter monsoon vs summer monsoon) than daily weather type
- **Action:** Consider removing wind variables unless specifically modeling wind itself

**Precipitation Amount (Œ∑¬≤ = 0.041)**
- Surprisingly weak despite strong precipprob correlation
- **Why:** "Rain" category includes:
  - Light drizzle: 0.5 mm
  - Moderate rain: 5-10 mm
  - Heavy storms: 30+ mm
- Intensity information is lost in categorical labeling

**Visibility (Œ∑¬≤ = 0.061)**
- Weather conditions barely explain visibility (6% of variance)
- Other factors may be more important: air pollution (PM2.5), fog timing, local emissions
- Hanoi's visibility is often pollution-driven rather than weather-driven

---

**Feature Engineering**

**High-Value Categorical Predictors**
1. Keep `conditions` for predicting:
   - Precipitation probability (Œ∑¬≤ = 1.0)
   - Cloud cover (Œ∑¬≤ = 0.57)
   - Precipitation coverage (Œ∑¬≤ = 0.25)

**Low-Value Features to Remove**
1. **Wind variables** (Œ∑¬≤ < 0.03) - minimal association with any categorical feature
2. **Sunrise/sunset categorical strings** - spurious perfect correlation (Œ∑¬≤ = 1.0)
3. **Visibility** (Œ∑¬≤ = 0.06) - better predicted by pollution data if available

**Transform Categorical Time Features**
Replace sunrise/sunset strings with numerical features:
1. **Day length** (hours): `sunset_time - sunrise_time` ‚Üí captures seasonal variation (11.5-13.5 hours)
2. **Solar hour**: Convert time strings to decimal hours (6:45 AM ‚Üí 6.75)
3. **Seasonal encoding**: `sin(2œÄ √ó day_of_year / 365)` and `cos(2œÄ √ó day_of_year / 365)` ‚Üí captures cyclical patterns

**Create Interaction Features for Temperature**
Since conditions alone explain only 14-16% of temperature variance:
- `conditions √ó month`: Captures "Rain in July" vs "Rain in January"
- `conditions √ó day_of_year_sin/cos`: Continuous seasonal interaction
- Example: Model learns different temperature distributions for each condition-season combination
