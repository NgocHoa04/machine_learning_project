## Train - Test Split

```python
import pandas as pd


class TimeSeriesSplitter:
    """
    Clean train/val/test splitter for time-series forecasting
    """

    def __init__(self, datetime_col="datetime"):
        self.datetime_col = datetime_col

    def split(self, df: pd.DataFrame, train_ratio=0.7, val_ratio=0.15):
        df = df.sort_values(self.datetime_col).reset_index(drop=True)

        n = len(df)
        train_end = int(n * train_ratio)
        val_end = int(n * (train_ratio + val_ratio))

        train = df.iloc[:train_end]
        val = df.iloc[train_end:val_end]
        test = df.iloc[val_end:]

        return train, val, test
```

```python
train, val, test = TimeSeriesSplitter(datetime_col="datetime").split(hn_hourly, train_ratio=0.8)
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.feature_selection import VarianceThreshold
from sklearn.pipeline import Pipeline
import numpy as np

class VarianceThresholdSelector(BaseEstimator, TransformerMixin):
    def __init__(self, threshold=0):
        self.threshold = threshold
        self.selector = VarianceThreshold(threshold=self.threshold)
    
    def fit(self, X, y=None):
        numeric_cols = X.select_dtypes(include=['float64', 'int64']).columns
        self.numeric_cols = numeric_cols
        
        self.selector.fit(X[numeric_cols])

        self.retained_cols = list(numeric_cols[self.selector.get_support()])
        self.dropped_cols = [col for col in numeric_cols if col not in self.retained_cols]

        print("C√°c c·ªôt b·ªã lo·∫°i v√¨ ph∆∞∆°ng sai = 0:", self.dropped_cols)
        return self
    
    def transform(self, X):
        cols_to_keep = self.retained_cols + list(X.columns.difference(self.numeric_cols))
        self.output_cols_ = cols_to_keep  # <- l∆∞u l·∫°i ƒë·ªÉ get_feature_names_out
        return X[cols_to_keep]

    def get_feature_names_out(self, input_features=None):
        return np.array(self.output_cols_)


class ConstantAndDuplicateRemover(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        self.constant_cols = [col for col in X.columns if X[col].nunique() == 1]
        print("C√°c c·ªôt b·ªã lo·∫°i v√¨ ch·ªâ c√≥ 1 gi√° tr·ªã:", self.constant_cols)

        # L∆∞u l·∫°i danh s√°ch c·ªôt sau khi lo·∫°i
        self.output_cols_ = [col for col in X.columns if col not in self.constant_cols]
        return self
    
    def transform(self, X):
        X_cleaned = X.drop(columns=self.constant_cols, errors="ignore")
        X_cleaned = X_cleaned.drop_duplicates()
        return X_cleaned

    def get_feature_names_out(self, input_features=None):
        return np.array(self.output_cols_)
```

## Correlation Analysis for Categorical and Mixed Variables

### 1. Categorical-Categorical Correlation: Cram√©r's V
Cram√©r's V ƒëo l∆∞·ªùng m·ªëi quan h·ªá gi·ªØa 2 bi·∫øn categorical d·ª±a tr√™n chi-square test. Gi√° tr·ªã t·ª´ 0 (kh√¥ng c√≥ m·ªëi quan h·ªá) ƒë·∫øn 1 (m·ªëi quan h·ªá ho√†n h·∫£o).

### 2. Numerical-Categorical Correlation: Correlation Ratio (Œ∑¬≤)
Correlation ratio (eta-squared) ƒëo l∆∞·ªùng m·ª©c ƒë·ªô m·ªôt bi·∫øn categorical gi·∫£i th√≠ch ƒë∆∞·ª£c s·ª± bi·∫øn thi√™n c·ªßa m·ªôt bi·∫øn numerical. Gi√° tr·ªã t·ª´ 0 ƒë·∫øn 1.

```python
# Import th∆∞ vi·ªán c·∫ßn thi·∫øt
import numpy as np
from scipy.stats import chi2_contingency
import warnings
warnings.filterwarnings('ignore')

# H√†m t√≠nh Cram√©r's V cho categorical-categorical correlation
def cramers_v(x, y):
    """
    T√≠nh Cram√©r's V gi·ªØa 2 bi·∫øn categorical
    V = sqrt(chi2 / (n * (min(k,r) - 1)))
    """
    # Lo·∫°i b·ªè missing values
    mask = ~(pd.isna(x) | pd.isna(y))
    x_clean = x[mask]
    y_clean = y[mask]
    
    if len(x_clean) == 0:
        return np.nan
    
    # T·∫°o contingency table
    confusion_matrix = pd.crosstab(x_clean, y_clean)
    
    # Chi-square test
    chi2, _, _, _ = chi2_contingency(confusion_matrix)
    
    n = confusion_matrix.sum().sum()
    min_dim = min(confusion_matrix.shape[0], confusion_matrix.shape[1]) - 1
    
    if min_dim == 0:
        return 0
    
    v = np.sqrt(chi2 / (n * min_dim))
    return v

# H√†m t√≠nh correlation ratio (eta-squared) cho numerical-categorical correlation
def correlation_ratio(categories, values):
    """
    T√≠nh correlation ratio (Œ∑¬≤) gi·ªØa bi·∫øn categorical v√† numerical
    Œ∑¬≤ = SSB / SST (Sum of Squares Between / Sum of Squares Total)
    """
    # Lo·∫°i b·ªè missing values
    mask = ~(pd.isna(categories) | pd.isna(values))
    categories_clean = categories[mask]
    values_clean = values[mask]
    
    if len(values_clean) == 0:
        return np.nan
    
    # T√≠nh mean t·ªïng th·ªÉ
    mean_total = values_clean.mean()
    
    # T√≠nh SST (Total Sum of Squares)
    sst = np.sum((values_clean - mean_total) ** 2)
    
    if sst == 0:
        return 0
    
    # T√≠nh SSB (Between group Sum of Squares)
    ssb = 0
    for category in categories_clean.unique():
        group_values = values_clean[categories_clean == category]
        group_mean = group_values.mean()
        group_size = len(group_values)
        ssb += group_size * (group_mean - mean_total) ** 2
    
    # Œ∑¬≤ = SSB / SST
    eta_squared = ssb / sst
    return eta_squared

print("‚úÖ ƒê√£ ƒë·ªãnh nghƒ©a c√°c h√†m t√≠nh correlation")
```

### Outlier Handling Rationale
We will clipping instead of removal. Because:

1. **Temporal Dependency Preservation**
   - Time series requires continuous sequential data
   - Removing observations creates gaps that break temporal patterns
   - Models like LSTM/GRU need uninterrupted sequences

2. **Physical Reality**
   - Outliers represent extreme weather events (storms, heatwaves)
   - These are valid measurements, not errors
   - Removal loses critical information about climate variability

3. **Domain-Based Boundaries**
   - Physical constraints are well-defined (e.g., humidity: 0-100%)
   - Values exceeding limits are measurement errors, not natural phenomena
   - Clipping corrects sensor errors while preserving temporal continuity

4. **Statistical Considerations**
   - Small dataset (~8,000 hourly records) - cannot afford data loss
   - Outliers may be rare but informative for prediction
   - Clipping maintains sample size for model training

```python
class OutlierHandler(BaseEstimator, TransformerMixin):
    """
    X·ª≠ l√Ω outliers d·ª±a tr√™n domain knowledge v√† IQR method
    """
    def __init__(self, method='clip'):
        self.method = method
        self.bounds = {}
    
    def fit(self, X, y=None):
        """H·ªçc boundaries t·ª´ training data"""
        for col in X.select_dtypes(include=[np.number]).columns:
            Q1 = X[col].quantile(0.25)
            Q3 = X[col].quantile(0.75)
            IQR = Q3 - Q1
            
            # Physical constraints cho t·ª´ng bi·∫øn
            if col == 'solarradiation':
                # Max solar radiation ~1361 W/m¬≤ (solar constant)
                lower = 0
                upper = 1400
            elif col == 'precip':
                # L∆∞·ª£ng m∆∞a kh√¥ng √¢m, max ~50mm/hour (extreme)
                lower = 0
                upper = 50
            elif col == 'winddir':
                # Wind direction: 0-360¬∞
                lower = 0
                upper = 360
            elif col == 'sealevelpressure':
                # Pressure: 950-1050 hPa (reasonable range)
                lower = 950
                upper = 1050
            elif col == 'humidity':
                # Humidity: 0-100%
                lower = 0
                upper = 100
            elif col == 'windspeed' or col == 'windgust':
                # Wind speed: kh√¥ng √¢m, max ~50 m/s
                lower = 0
                upper = X[col].quantile(0.99)  # 99th percentile
            else:
                # IQR method cho c√°c bi·∫øn kh√°c
                lower = Q1 - 1.5 * IQR
                upper = Q3 + 1.5 * IQR
            
            self.bounds[col] = (lower, upper)
        
        return self
    
    def transform(self, X):
        """Apply outlier handling"""
        X = X.copy()
        
        for col, (lower, upper) in self.bounds.items():
            if col in X.columns:
                if self.method == 'clip':
                    # Clip outliers to boundaries
                    X[col] = X[col].clip(lower=lower, upper=upper)
                elif self.method == 'winsorize':
                    # Winsorize at boundaries
                    X[col] = np.where(X[col] < lower, lower, X[col])
                    X[col] = np.where(X[col] > upper, upper, X[col])
        
        return X
```

### Missing Value Imputation Strategy

**1. Drop: severerisk (81.7% missing)**
- Excessive missing rate leads to unreliable imputation and potential bias

**2. Forward/Backward Fill: visibility, windspeed, winddir, sealevelpressure**
- Meteorological variables exhibit temporal continuity (slow-changing patterns)
- Forward fill assumes persistence over short intervals; backward fill handles series start

**3. Conditional Imputation: precip (0.05% missing)**
- Domain logic: precipprob = 0 ‚Üí precip = 0 (no rain expected)
- Remaining cases: median imputation (robust to outliers in skewed distribution)

**4. Time-based + Interpolation: solarradiation, solarenergy, uvindex (~0.046% missing)**
- Physical constraint: nighttime or heavy cloud cover (‚â•90%) ‚Üí solar = 0
- Daytime gaps: linear interpolation leverages diurnal sinusoidal pattern

**5. KNN Imputation: windgust (0.06% missing)**
- High correlation between wind variables (speed, direction, gust)
- KNN captures multivariate relationships using k=5 neighbors with distance weighting

```python
from sklearn.impute import SimpleImputer, KNNImputer

class MissingValueHandlerHourly(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.knn_imputer = KNNImputer(n_neighbors=5, weights='distance')
        self.medians = {}
    
    def fit(self, X, y=None):
        # L∆∞u median values
        for col in X.select_dtypes(include=[np.number]).columns:
            self.medians[col] = X[col].median()
        
        # Fit KNN imputer
        wind_cols = ['windspeed', 'winddir', 'windgust']
        if all(col in X.columns for col in wind_cols):
            self.knn_imputer.fit(X[wind_cols])
        
        return self
    
    def transform(self, X):
        X = X.copy()
        
        # 1. Drop severerisk
        X.drop(columns=['severerisk'], inplace=True, errors='ignore')
        
        # 2. Forward fill time series
        time_cols = ['visibility', 'windspeed', 'winddir', 'sealevelpressure']
        for col in time_cols:
            if col in X.columns:
                X[col] = X[col].fillna(method='ffill').fillna(method='bfill')
        
        # 3. Conditional precip
        if 'precip' in X.columns:
            X.loc[X['precip'].isna() & (X['precipprob'] == 0), 'precip'] = 0
            X['precip'].fillna(self.medians.get('precip', 0), inplace=True)
        
        # 4. Solar features
        solar_cols = ['solarradiation', 'solarenergy', 'uvindex']
        for col in solar_cols:
            if col in X.columns:
                X.loc[(X[col].isna()) & (X.get('is_night', 0) == 1), col] = 0
                X[col] = X[col].interpolate(method='linear').fillna(0)
        
        # 5. KNN for wind
        wind_cols = ['windspeed', 'winddir', 'windgust']
        if all(col in X.columns for col in wind_cols):
            X[wind_cols] = self.knn_imputer.transform(X[wind_cols])
        
        return X
```

```python
# S·ª≠ d·ª•ng MissingValueHandlerHourly ƒë·ªÉ x·ª≠ l√Ω missing values v·ªõi X_train ƒë√£ l√†m s·∫°ch
mv_handler = MissingValueHandlerHourly()
X_train_imputed = mv_handler.fit_transform(X_train_after_outlier.copy())
missing_after = X_train_imputed.isna().sum().sort_values(ascending=False)
missing_pct_after = (X_train_imputed.isna().mean().sort_values(ascending=False))*100
pd.concat([missing_after, missing_pct_after], axis=1, keys=["missing_after","percent_after"])
```
