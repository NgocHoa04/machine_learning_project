```python

import pandas as pd
import numpy as np
import xgboost as xgb
import optuna
from sklearn.metrics import mean_squared_error

class WalkForwardOptunaPipeline:
    def __init__(self, df, date_col, target_col, feature_cols, n_splits=5, test_size=30, mode="expanding"):
        """
        df          : pandas DataFrame theo thứ tự thời gian
        date_col    : tên cột ngày tháng (datetime)
        target_col  : tên cột target (vd: 'temp')
        feature_cols: list các cột feature (KHÔNG nên chứa target)
        n_splits    : số fold walk-forward
        test_size   : số ngày test cho mỗi fold
        mode        : "expanding" hoặc "rolling"
        """
        self.df = df.copy().sort_values(date_col).reset_index(drop=True)
        self.date_col = date_col
        self.target_col = target_col
        self.feature_cols = feature_cols
        self.n_splits = n_splits
        self.test_size = test_size
        self.mode = mode
        self.walkfolds = []
        self.final_model = None
        self.best_params = None

        # Feature thực sự dùng train: loại bỏ date/target nếu lỡ nằm trong feature_cols
        self.X_cols = [
            c for c in self.feature_cols
            if c not in [
                self.date_col,
                self.target_col,
                self.target_col + "_next",
                self.date_col + "_next"
            ]
        ]

    def add_target_shift(self):
        """Shift target sang ngày mai"""
        self.df[self.target_col + "_next"] = self.df[self.target_col].shift(-1)
        self.df[self.date_col + "_next"] = self.df[self.date_col].shift(-1)
        self.df = self.df.dropna().reset_index(drop=True)

    def create_walkforward_folds(self):
        """Tạo các fold train/test dạng: ((X_train, y_train), (X_val, y_val))"""
        df_len = len(self.df)
        step = (df_len - self.test_size) // self.n_splits
        folds = []

        for i in range(self.n_splits):
            train_end = step * (i + 1)
            test_start = train_end + 91   # bỏ gap 91 ngày
            test_end = test_start + self.test_size
            if test_end > df_len:
                break

            if self.mode == "expanding":
                train_start = 0
            else:  # rolling
                train_start = max(0, train_end - step * 2)

            train_df = self.df.iloc[train_start:train_end]
            test_df  = self.df.iloc[test_start:test_end]

            X_train = train_df[self.X_cols]
            y_train = train_df[self.target_col + "_next"]

            X_val = test_df[self.X_cols]
            y_val = test_df[self.target_col + "_next"]

            folds.append(
                ((X_train.reset_index(drop=True), y_train.reset_index(drop=True)),
                 (X_val.reset_index(drop=True),   y_val.reset_index(drop=True)))
            )

        self.walkfolds = folds

    def create_objective(self):
        """Closure cho Optuna, dùng walk-forward folds"""
        def objective(trial):
            params = {
                'objective': 'reg:squarederror',
                'eval_metric': 'rmse',   # đặt trong model, không truyền vào fit
                'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
                'max_depth': trial.suggest_int('max_depth', 3, 10),
                'gamma': trial.suggest_float('gamma', 0.0, 1.0),
                'reg_lambda': trial.suggest_float('reg_lambda', 1.0, 10.0),
                'reg_alpha': trial.suggest_float('reg_alpha', 0.0, 10.0),
                'subsample': trial.suggest_float('subsample', 0.5, 1.0),
                'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
            }

            scores = []
            for (X_train, y_train), (X_val, y_val) in self.walkfolds:
                model = xgb.XGBRegressor(
                    **params,
                    n_estimators=800,   # giảm bớt cho nhanh, cần thì tăng
                    random_state=42,
                    tree_method='hist'
                )

                # fit đơn giản, không callbacks, không early stopping
                model.fit(
                    X_train,
                    y_train,
                    eval_set=[(X_val, y_val)],
                    verbose=False
                )

                preds = model.predict(X_val)
                rmse = mean_squared_error(y_val, preds)
                scores.append(rmse)

            return np.mean(scores)
        return objective

    def run_optuna(self, n_trials=50):
        """Chạy Optuna tìm hyperparameters"""
        objective_fn = self.create_objective()
        study = optuna.create_study(direction='minimize')
        study.optimize(objective_fn, n_trials=n_trials)
        self.best_params = study.best_params
        return study

    def train_final_model(self):
        """Huấn luyện final model trên toàn bộ development set"""
        if self.best_params is None:
            raise ValueError("Bạn phải chạy run_optuna() trước khi train_final_model().")

        final_params = self.best_params.copy()
        final_params.update({
            'objective': 'reg:squarederror',
            'eval_metric': 'rmse',
            'n_estimators': 800   # sync với phía trên
        })

        self.final_model = xgb.XGBRegressor(
            **final_params,
            random_state=42,
            tree_method='hist'
        )

        X_full = self.df[self.X_cols]
        y_full = self.df[self.target_col + "_next"]

        # fit full, không callbacks, không early stopping
        self.final_model.fit(
            X_full,
            y_full,
            verbose=False
        )

    def predict_next_day(self, X_today):
        """Dự đoán cho ngày mai, X_today là dataframe 1 dòng với đủ self.X_cols"""
        if self.final_model is None:
            raise ValueError("Bạn phải train final_model trước")
        X_input = X_today[self.X_cols]
        return self.final_model.predict(X_input)
```
