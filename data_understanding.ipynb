### **solarradiation: bá»©c xáº¡ máº·t trá»i táº¡i máº·t Ä‘áº¥t**

### An overview:
- Táº¡i máº·t Ä‘áº¥t, â€œsolarradiationâ€ Ä‘o cÃ´ng suáº¥t bá»©c xáº¡ (irradiance) tá»©c lÃ  nÄƒng lÆ°á»£ng (Watt) tá»›i má»™t Ä‘Æ¡n vá»‹ diá»‡n tÃ­ch (W/mÂ²) táº¡i má»™t thá»i Ä‘iá»ƒm hoáº·c trung bÃ¬nh cá»§a khoáº£ng thá»i gian.
- ÄÆ¡n vá»‹: W/mÂ² (Watt trÃªn mÃ©t vuÃ´ng) â€“ vá»›i há»‡ Ä‘o lÆ°á»ng metric.
- Náº¿u dá»¯ liá»‡u dÆ°á»›i dáº¡ng "daily", thÃ¬ con sá»‘ cá»§a solarradiation lÃ  giÃ¡ trá»‹ trung bÃ¬nh cá»§a cÃ¡c giÃ¡ trá»‹ hourly trong ngÃ y.

### Meaning:
- Biáº¿n nÃ y cho biáº¿t má»©c Ä‘á»™ â€œnhiá»‡t/Ã¡nh sÃ¡ng máº·t trá»iâ€ tá»›i bá» máº·t táº¡i thá»i Ä‘iá»ƒm hoáº·c trung bÃ¬nh giá» â€” cÃ ng cao nghÄ©a lÃ  cÃ³ Ã¡nh náº¯ng máº¡nh, Ã­t mÃ¢y, tia máº·t trá»i chiáº¿u gáº§n trá»±c tiáº¿p.
- Ráº¥t há»¯u Ã­ch Ä‘á»ƒ phÃ¢n tÃ­ch áº£nh hÆ°á»Ÿng thá»i tiáº¿t (mÃ¢y, mÆ°a, gÃ³c máº·t trá»i) tá»›i bá»©c xáº¡. VÃ­ dá»¥: khi mÃ¢y che nhiá»u â†’ solarradiation tháº¥p, káº¿t há»£p vá»›i cÃ¡c biáº¿n khÃ¡c (cloudcover, sun elevation, tilt angle) Ä‘á»ƒ mÃ´ hÃ¬nh hÃ³a hiá»‡u suáº¥t thá»±c táº¿.
- VÃ¬ lÃ  cÃ´ng suáº¥t/Ä‘Æ¡n vá»‹ diá»‡n tÃ­ch táº¡i má»™t thá»i Ä‘iá»ƒm (hoáº·c trung bÃ¬nh giá»), nÃªn náº¿u muá»‘n biáº¿t â€œnÄƒng lÆ°á»£ng tÃ­ch lÅ©yâ€ trong má»™t khoáº£ng thá»i gian (giá»/ngÃ y) thÃ¬ cáº§n Ä‘áº¿n biáº¿n solarenergy.

### Note:
- Dá»¯ liá»‡u giá» (hourly) sáº½ cÃ³ biáº¿n solarradiation má»—i giá» (hoáº·c trung bÃ¬nh giá»), náº¿u chá»‰ láº¥y giÃ¡ trá»‹ ngÃ y sáº½ lÃ  trung bÃ¬nh ngÃ y, máº¥t Ä‘i chi tiáº¿t dao Ä‘á»™ng trong ngÃ y.
- GiÃ¡ trá»‹ ráº¥t phá»¥ thuá»™c vÃ o: vá»‹ trÃ­ Ä‘á»‹a lÃ½ (vÄ© Ä‘á»™, kinh Ä‘á»™), hÆ°á»›ng bá» máº·t, Ä‘á»™ che phá»§ mÃ¢y, gÃ³c chiáº¿u máº·t trá»i (sun elevation), tÃ¬nh tráº¡ng khÃ´ng khÃ­ (haze, bá»¥i)
- Khi sá»­ dá»¥ng cho má»¥c Ä‘Ã­ch mÃ´ hÃ¬nh hÃ³a nÄƒng lÆ°á»£ng hoáº·c so sÃ¡nh giá»¯a cÃ¡c ngÃ y, cáº§n lÆ°u Ã½ Ä‘Æ¡n vá»‹ W/mÂ² vÃ  chuyá»ƒn Ä‘á»•i náº¿u cáº§n.
- Biáº¿n nÃ y khÃ´ng cho biáº¿t tá»•ng nÄƒng lÆ°á»£ng tÃ­ch lÅ©y trong ngÃ y â€” chá»‰ lÃ  má»©c â€œtáº¡i thá»i Ä‘iá»ƒmâ€ hoáº·c trung bÃ¬nh giá» â€” nÃªn náº¿u muá»‘n tá»•ng nÄƒng lÆ°á»£ng, dÃ¹ng solarenergy.

````python
import matplotlib.pyplot as plt
import pandas as pd

# Chuyá»ƒn Ä‘á»•i cá»™t datetime sang dáº¡ng datetime
hanoi_daily_df['datetime'] = pd.to_datetime(hanoi_daily_df['datetime'])
plt.figure(figsize=(16, 6))
plt.plot(hanoi_daily_df['datetime'], hanoi_daily_df['solarradiation'], 
         label='Solar Radiation (W/mÂ²)', color='orange', linewidth=1.5, alpha=0.8)
plt.plot(hanoi_daily_df['datetime'], hanoi_daily_df['temp'], 
         label='Temperature (Â°C)', color='blue', linewidth=1.5, alpha=0.8)

# Thiáº¿t láº­p labels vÃ  title
plt.xlabel('Date', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.title('Solar Radiation and Temperature over time', fontsize=14, fontweight='bold')

plt.grid(True, alpha=0.3, linestyle='--')
plt.legend(loc='upper right', fontsize=10)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# In thá»‘ng kÃª cÆ¡ báº£n
print("ðŸ“Š THá»NG KÃŠ CÆ  Báº¢N:")
print(f"Solar Radiation - Mean: {hanoi_daily_df['solarradiation'].mean():.2f} W/mÂ²")
print(f"Solar Radiation - Max: {hanoi_daily_df['solarradiation'].max():.2f} W/mÂ²")
print(f"Solar Radiation - Min: {hanoi_daily_df['solarradiation'].min():.2f} W/mÂ²")
print(f"\nTemperature - Mean: {hanoi_daily_df['temp'].mean():.2f} Â°C")
print(f"Temperature - Max: {hanoi_daily_df['temp'].max():.2f} Â°C")
print(f"Temperature - Min: {hanoi_daily_df['temp'].min():.2f} Â°C")
print(f"\nCorrelation: {hanoi_daily_df['solarradiation'].corr(hanoi_daily_df['temp']):.3f}")
````

````python
# Váº½ scatter plot Ä‘á»ƒ xem má»‘i quan há»‡ trá»±c tiáº¿p giá»¯a solarradiation vÃ  temperature
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Subplot 1: Scatter plot vá»›i regression line
ax1.scatter(hanoi_daily_df['solarradiation'], hanoi_daily_df['temp'], 
           alpha=0.5, color='coral', edgecolors='black', linewidth=0.5)

# ThÃªm regression line
z = np.polyfit(hanoi_daily_df['solarradiation'], hanoi_daily_df['temp'], 1)
p = np.poly1d(z)
ax1.plot(hanoi_daily_df['solarradiation'], p(hanoi_daily_df['solarradiation']), 
         "r--", linewidth=2, label=f'Regression line: y={z[0]:.4f}x+{z[1]:.2f}')

ax1.set_xlabel('Solar Radiation (W/mÂ²)', fontsize=12)
ax1.set_ylabel('Temperature (Â°C)', fontsize=12)
ax1.set_title('Relationship: Solar Radiation vs Temperature', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Subplot 2: Box plot Ä‘á»ƒ so sÃ¡nh distribution
data_to_plot = [hanoi_daily_df['solarradiation'], hanoi_daily_df['temp']]
ax2.boxplot(data_to_plot, labels=['Solar Radiation\n(W/mÂ²)', 'Temperature\n(Â°C)'], 
            patch_artist=True, 
            boxprops=dict(facecolor='lightblue', alpha=0.7),
            medianprops=dict(color='red', linewidth=2))

ax2.set_ylabel('Value', fontsize=12)
ax2.set_title('Distribution Comparison', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

# PhÃ¢n tÃ­ch má»‘i quan há»‡
corr = hanoi_daily_df['solarradiation'].corr(hanoi_daily_df['temp'])
print(f"Há»‡ sá»‘ tÆ°Æ¡ng quan (Correlation): {corr:.3f}")
    
print(f"\nðŸ’¡ Khi Solar Radiation tÄƒng 1 W/mÂ², Temperature tÄƒng khoáº£ng {z[0]:.4f}Â°C")
````
