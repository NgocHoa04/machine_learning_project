## Basic Processing

````python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.feature_selection import VarianceThreshold
from sklearn.pipeline import Pipeline
import numpy as np

# B∆∞·ªõc 1: Lo·∫°i b·ªè c·ªôt s·ªë c√≥ ph∆∞∆°ng sai = 0
class VarianceThresholdSelector(BaseEstimator, TransformerMixin):
    def __init__(self, threshold=0):
        self.threshold = threshold
        self.selector = VarianceThreshold(threshold=self.threshold)
    
    def fit(self, X, y=None):
        numeric_cols = X.select_dtypes(include=['float64', 'int64']).columns
        self.numeric_cols = numeric_cols
        self.selector.fit(X[numeric_cols])
        # L∆∞u l·∫°i c·ªôt s·ªë ƒë∆∞·ª£c gi·ªØ l·∫°i
        self.retained_cols = numeric_cols[self.selector.get_support()]
        # L∆∞u l·∫°i c·ªôt s·ªë b·ªã lo·∫°i
        self.dropped_cols = [col for col in numeric_cols if col not in self.retained_cols]
        print("C√°c c·ªôt b·ªã lo·∫°i v√¨ ph∆∞∆°ng sai = 0:", self.dropped_cols)
        return self
    
    def transform(self, X):
        cols_to_keep = list(self.retained_cols) + list(X.columns.difference(self.numeric_cols))
        return X[cols_to_keep]

# B∆∞·ªõc 2: Lo·∫°i b·ªè c·ªôt h·∫±ng s·ªë v√† d√≤ng tr√πng l·∫∑p
class ConstantAndDuplicateRemover(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        # X√°c ƒë·ªãnh c·ªôt h·∫±ng s·ªë
        self.constant_cols = [col for col in X.columns if X[col].nunique() == 1]
        print("C√°c c·ªôt b·ªã lo·∫°i v√¨ ch·ªâ c√≥ 1 gi√° tr·ªã:", self.constant_cols)
        return self
    
    def transform(self, X):
        # B·ªè c·ªôt h·∫±ng s·ªë
        X_cleaned = X.drop(columns=self.constant_cols, errors="ignore")
        # B·ªè d√≤ng tr√πng l·∫∑p (kh√¥ng in s·ªë l∆∞·ª£ng n·ªØa)
        X_cleaned = X_cleaned.drop_duplicates()
        return X_cleaned

    
# T·∫°o pipeline
cleaning_pipeline = Pipeline(steps=[
    ("var_thresh", VarianceThresholdSelector(threshold=0)),
    ("const_dup", ConstantAndDuplicateRemover())

])
X_cleaned = cleaning_pipeline.fit_transform(X_train.copy())
print(f"K√≠ch th∆∞·ªõc d·ªØ li·ªáu sau khi l√†m s·∫°ch: {X_cleaned.shape}")
````

## Correlation Analysis for Categorical and Mixed Variables

### 1. Categorical-Categorical Correlation: Cram√©r's V
Cram√©r's V ƒëo l∆∞·ªùng m·ªëi quan h·ªá gi·ªØa 2 bi·∫øn categorical d·ª±a tr√™n chi-square test. Gi√° tr·ªã t·ª´ 0 (kh√¥ng c√≥ m·ªëi quan h·ªá) ƒë·∫øn 1 (m·ªëi quan h·ªá ho√†n h·∫£o).

### 2. Numerical-Categorical Correlation: Correlation Ratio (Œ∑¬≤)
Correlation ratio (eta-squared) ƒëo l∆∞·ªùng m·ª©c ƒë·ªô m·ªôt bi·∫øn categorical gi·∫£i th√≠ch ƒë∆∞·ª£c s·ª± bi·∫øn thi√™n c·ªßa m·ªôt bi·∫øn numerical. Gi√° tr·ªã t·ª´ 0 ƒë·∫øn 1.

````python
# Import th∆∞ vi·ªán c·∫ßn thi·∫øt
import numpy as np
from scipy.stats import chi2_contingency
import warnings
warnings.filterwarnings('ignore')

# H√†m t√≠nh Cram√©r's V cho categorical-categorical correlation
def cramers_v(x, y):
    """
    T√≠nh Cram√©r's V gi·ªØa 2 bi·∫øn categorical
    V = sqrt(chi2 / (n * (min(k,r) - 1)))
    """
    # Lo·∫°i b·ªè missing values
    mask = ~(pd.isna(x) | pd.isna(y))
    x_clean = x[mask]
    y_clean = y[mask]
    
    if len(x_clean) == 0:
        return np.nan
    
    # T·∫°o contingency table
    confusion_matrix = pd.crosstab(x_clean, y_clean)
    
    # Chi-square test
    chi2, _, _, _ = chi2_contingency(confusion_matrix)
    
    n = confusion_matrix.sum().sum()
    min_dim = min(confusion_matrix.shape[0], confusion_matrix.shape[1]) - 1
    
    if min_dim == 0:
        return 0
    
    v = np.sqrt(chi2 / (n * min_dim))
    return v

# H√†m t√≠nh correlation ratio (eta-squared) cho numerical-categorical correlation
def correlation_ratio(categories, values):
    """
    T√≠nh correlation ratio (Œ∑¬≤) gi·ªØa bi·∫øn categorical v√† numerical
    Œ∑¬≤ = SSB / SST (Sum of Squares Between / Sum of Squares Total)
    """
    # Lo·∫°i b·ªè missing values
    mask = ~(pd.isna(categories) | pd.isna(values))
    categories_clean = categories[mask]
    values_clean = values[mask]
    
    if len(values_clean) == 0:
        return np.nan
    
    # T√≠nh mean t·ªïng th·ªÉ
    mean_total = values_clean.mean()
    
    # T√≠nh SST (Total Sum of Squares)
    sst = np.sum((values_clean - mean_total) ** 2)
    
    if sst == 0:
        return 0
    
    # T√≠nh SSB (Between group Sum of Squares)
    ssb = 0
    for category in categories_clean.unique():
        group_values = values_clean[categories_clean == category]
        group_mean = group_values.mean()
        group_size = len(group_values)
        ssb += group_size * (group_mean - mean_total) ** 2
    
    # Œ∑¬≤ = SSB / SST
    eta_squared = ssb / sst
    return eta_squared

print("‚úÖ ƒê√£ ƒë·ªãnh nghƒ©a c√°c h√†m t√≠nh correlation:")
````

### T√≠nh Categorical-Categorical Correlation (Cram√©r's V)

````python
# C·∫≠p nh·∫≠t l·∫°i categorical_cols ƒë·ªÉ ch·ªâ gi·ªØ c√°c c·ªôt c√≤n t·ªìn t·∫°i trong X_cleaned
categorical_cols_cleaned = [col for col in categorical_cols if col in X_cleaned.columns]

# T√≠nh Cram√©r's V matrix cho t·∫•t c·∫£ c√°c c·∫∑p bi·∫øn categorical c√≤n l·∫°i
if len(categorical_cols_cleaned) > 0:
    cat_cat_corr = pd.DataFrame(
        np.zeros((len(categorical_cols_cleaned), len(categorical_cols_cleaned))),
        index=categorical_cols_cleaned,
        columns=categorical_cols_cleaned
    )

    for i, col1 in enumerate(categorical_cols_cleaned):
        for j, col2 in enumerate(categorical_cols_cleaned):
            if i <= j:  # Ch·ªâ t√≠nh n·ª≠a tr√™n c·ªßa ma tr·∫≠n (symmetric)
                v = cramers_v(X_cleaned[col1], X_cleaned[col2])
                cat_cat_corr.loc[col1, col2] = v
                cat_cat_corr.loc[col2, col1] = v

    print("üìä Cram√©r's V Matrix (Categorical-Categorical):")
    display(cat_cat_corr)
else:
    print("‚ö†Ô∏è Kh√¥ng c√≤n c·ªôt categorical n√†o trong X_cleaned ƒë·ªÉ ph√¢n t√≠ch!")
````

````python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, FunctionTransformer, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

# Custom transformer: log1p cho c√°c c·ªôt outlier
log_transformer = FunctionTransformer(np.log1p, validate=True)

# C√°c nh√≥m c·ªôt
outlier_cols = ["precip", "solarradiation"]
scale_cols = [ "windspeed"]
minmax_cols = ["precipprob", "cloudcover", "precipcover"]
humidity_col = ["humidity"]

humidity_scale_function = FunctionTransformer(lambda x: x / 100.0, validate=True)

# Pipeline cho c√°c c·ªôt outlier
log_outlier_pipeline = Pipeline(steps=[
    ('log_trans', log_transformer) # Bi·∫øn ƒë·ªïi log(x+1)
])
````
