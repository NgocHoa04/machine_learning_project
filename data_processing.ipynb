## Basic Processing

````python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.feature_selection import VarianceThreshold
from sklearn.pipeline import Pipeline
import numpy as np

# B∆∞·ªõc 1: Lo·∫°i b·ªè c·ªôt s·ªë c√≥ ph∆∞∆°ng sai = 0
class VarianceThresholdSelector(BaseEstimator, TransformerMixin):
    def __init__(self, threshold=0):
        self.threshold = threshold
        self.selector = VarianceThreshold(threshold=self.threshold)
    
    def fit(self, X, y=None):
        numeric_cols = X.select_dtypes(include=['float64', 'int64']).columns
        self.numeric_cols = numeric_cols
        self.selector.fit(X[numeric_cols])
        # L∆∞u l·∫°i c·ªôt s·ªë ƒë∆∞·ª£c gi·ªØ l·∫°i
        self.retained_cols = numeric_cols[self.selector.get_support()]
        # L∆∞u l·∫°i c·ªôt s·ªë b·ªã lo·∫°i
        self.dropped_cols = [col for col in numeric_cols if col not in self.retained_cols]
        print("C√°c c·ªôt b·ªã lo·∫°i v√¨ ph∆∞∆°ng sai = 0:", self.dropped_cols)
        return self
    
    def transform(self, X):
        cols_to_keep = list(self.retained_cols) + list(X.columns.difference(self.numeric_cols))
        return X[cols_to_keep]

# B∆∞·ªõc 2: Lo·∫°i b·ªè c·ªôt h·∫±ng s·ªë v√† d√≤ng tr√πng l·∫∑p
class ConstantAndDuplicateRemover(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        # X√°c ƒë·ªãnh c·ªôt h·∫±ng s·ªë
        self.constant_cols = [col for col in X.columns if X[col].nunique() == 1]
        print("C√°c c·ªôt b·ªã lo·∫°i v√¨ ch·ªâ c√≥ 1 gi√° tr·ªã:", self.constant_cols)
        return self
    
    def transform(self, X):
        # B·ªè c·ªôt h·∫±ng s·ªë
        X_cleaned = X.drop(columns=self.constant_cols, errors="ignore")
        # B·ªè d√≤ng tr√πng l·∫∑p (kh√¥ng in s·ªë l∆∞·ª£ng n·ªØa)
        X_cleaned = X_cleaned.drop_duplicates()
        return X_cleaned

    
# T·∫°o pipeline
cleaning_pipeline = Pipeline(steps=[
    ("var_thresh", VarianceThresholdSelector(threshold=0)),
    ("const_dup", ConstantAndDuplicateRemover())

])
X_cleaned = cleaning_pipeline.fit_transform(X_train.copy())
print(f"K√≠ch th∆∞·ªõc d·ªØ li·ªáu sau khi l√†m s·∫°ch: {X_cleaned.shape}")
````

## Correlation Analysis for Categorical and Mixed Variables

### 1. Categorical-Categorical Correlation: Cram√©r's V
Cram√©r's V ƒëo l∆∞·ªùng m·ªëi quan h·ªá gi·ªØa 2 bi·∫øn categorical d·ª±a tr√™n chi-square test. Gi√° tr·ªã t·ª´ 0 (kh√¥ng c√≥ m·ªëi quan h·ªá) ƒë·∫øn 1 (m·ªëi quan h·ªá ho√†n h·∫£o).

### 2. Numerical-Categorical Correlation: Correlation Ratio (Œ∑¬≤)
Correlation ratio (eta-squared) ƒëo l∆∞·ªùng m·ª©c ƒë·ªô m·ªôt bi·∫øn categorical gi·∫£i th√≠ch ƒë∆∞·ª£c s·ª± bi·∫øn thi√™n c·ªßa m·ªôt bi·∫øn numerical. Gi√° tr·ªã t·ª´ 0 ƒë·∫øn 1.

````python
# Import th∆∞ vi·ªán c·∫ßn thi·∫øt
import numpy as np
from scipy.stats import chi2_contingency
import warnings
warnings.filterwarnings('ignore')

# H√†m t√≠nh Cram√©r's V cho categorical-categorical correlation
def cramers_v(x, y):
    """
    T√≠nh Cram√©r's V gi·ªØa 2 bi·∫øn categorical
    V = sqrt(chi2 / (n * (min(k,r) - 1)))
    """
    # Lo·∫°i b·ªè missing values
    mask = ~(pd.isna(x) | pd.isna(y))
    x_clean = x[mask]
    y_clean = y[mask]
    
    if len(x_clean) == 0:
        return np.nan
    
    # T·∫°o contingency table
    confusion_matrix = pd.crosstab(x_clean, y_clean)
    
    # Chi-square test
    chi2, _, _, _ = chi2_contingency(confusion_matrix)
    
    n = confusion_matrix.sum().sum()
    min_dim = min(confusion_matrix.shape[0], confusion_matrix.shape[1]) - 1
    
    if min_dim == 0:
        return 0
    
    v = np.sqrt(chi2 / (n * min_dim))
    return v

# H√†m t√≠nh correlation ratio (eta-squared) cho numerical-categorical correlation
def correlation_ratio(categories, values):
    """
    T√≠nh correlation ratio (Œ∑¬≤) gi·ªØa bi·∫øn categorical v√† numerical
    Œ∑¬≤ = SSB / SST (Sum of Squares Between / Sum of Squares Total)
    """
    # Lo·∫°i b·ªè missing values
    mask = ~(pd.isna(categories) | pd.isna(values))
    categories_clean = categories[mask]
    values_clean = values[mask]
    
    if len(values_clean) == 0:
        return np.nan
    
    # T√≠nh mean t·ªïng th·ªÉ
    mean_total = values_clean.mean()
    
    # T√≠nh SST (Total Sum of Squares)
    sst = np.sum((values_clean - mean_total) ** 2)
    
    if sst == 0:
        return 0
    
    # T√≠nh SSB (Between group Sum of Squares)
    ssb = 0
    for category in categories_clean.unique():
        group_values = values_clean[categories_clean == category]
        group_mean = group_values.mean()
        group_size = len(group_values)
        ssb += group_size * (group_mean - mean_total) ** 2
    
    # Œ∑¬≤ = SSB / SST
    eta_squared = ssb / sst
    return eta_squared

print("‚úÖ ƒê√£ ƒë·ªãnh nghƒ©a c√°c h√†m t√≠nh correlation")
````

````python
# C·∫≠p nh·∫≠t l·∫°i categorical_cols ƒë·ªÉ ch·ªâ gi·ªØ c√°c c·ªôt c√≤n t·ªìn t·∫°i trong X_cleaned
categorical_cols_cleaned = [col for col in categorical_cols if col in X_cleaned.columns]

# T√≠nh Cram√©r's V matrix cho t·∫•t c·∫£ c√°c c·∫∑p bi·∫øn categorical c√≤n l·∫°i
if len(categorical_cols_cleaned) > 0:
    cat_cat_corr = pd.DataFrame(
        np.zeros((len(categorical_cols_cleaned), len(categorical_cols_cleaned))),
        index=categorical_cols_cleaned,
        columns=categorical_cols_cleaned
    )

    for i, col1 in enumerate(categorical_cols_cleaned):
        for j, col2 in enumerate(categorical_cols_cleaned):
            if i <= j:  # Ch·ªâ t√≠nh n·ª≠a tr√™n c·ªßa ma tr·∫≠n (symmetric)
                v = cramers_v(X_cleaned[col1], X_cleaned[col2])
                cat_cat_corr.loc[col1, col2] = v
                cat_cat_corr.loc[col2, col1] = v

    print("üìä Cram√©r's V Matrix (Categorical-Categorical):")
    display(cat_cat_corr)
else:
    print("‚ö†Ô∏è Kh√¥ng c√≤n c·ªôt categorical n√†o trong X_cleaned ƒë·ªÉ ph√¢n t√≠ch!")
````

### üìä Cramer V Analysis: Categorical Variable Associations

**Overview**

Cramer V quantifies association strength between categorical variables through chi-square test statistics, ranging from 0 (complete independence) to 1 (perfect dependence). The analysis examines six categorical features: sunrise, sunset, conditions, description, icon, and stations, following removal of constant variables name and preciptype.

**Perfect Associations (V = 1.000): Critical Multicollinearity**

**Sunrise and Sunset Relationship**

Perfect correlation (V = 1.000) between sunrise and sunset arises from shared astronomical determinants. Both variables derive from identical inputs - geographic latitude (21 degrees North for Hanoi) and day of year - through deterministic celestial mechanics equations. Day length, computed as the difference between sunset and sunrise times, depends exclusively on day of year at fixed latitude. This mathematical dependency creates severe multicollinearity, necessitating removal of one variable to ensure model stability.

**Sunrise/Sunset and Weather Variables: Spurious Correlation**

Perfect associations (V = 1.000) between sunrise/sunset and weather variables (conditions, description, icon) represent methodological artifacts rather than meaningful relationships. Daily temporal granularity produces approximately 3,927 unique sunrise time categories (one per observation day), creating perfect one-to-one mapping between each sunrise string and corresponding weather condition. This spurious correlation occurs because sunrise times function as unique date identifiers rather than weather predictors, with no causal mechanism linking solar timing to atmospheric conditions.

The hidden confounder is date itself: each observation day determines both its unique sunrise time (via astronomical calculation) and weather condition (via atmospheric measurement). Future prediction scenarios will encounter unseen sunrise values, eliminating any generalization capacity. This represents pure memorization rather than pattern learning, causing catastrophic model overfitting if retained as features.

**Weather Descriptor Redundancy**

Near-perfect associations (V approximately 0.999-1.000) among conditions, description, and icon reflect intentional redundancy in weather API design. These three variables encode identical meteorological information in different formats: conditions provides concise labels (e.g., "Rain, Partially cloudy"), description offers narrative text (e.g., "Partly cloudy throughout the day with rain"), and icon supplies symbolic codes (e.g., "rain"). The weather service automatically generates all three representations from a single underlying classification. Retaining multiple encodings creates extreme multicollinearity, equivalent to learning the same feature multiple times. Recommended approach retains only conditions for its balance of conciseness and interpretability.

**Stations and Temporal Variables**

Perfect correlation (V = 1.000) between stations and sunrise/sunset times constitutes another spurious association mediated by date as hidden variable. The deterministic relationship flows through intermediate causation: date determines sunrise time (astronomical), date determines sunset time (astronomical), and date influences station combinations (data availability patterns). This three-way correlation lacks predictive value, as station metadata provides no information for forecasting solar timing.

**Weak Associations (V less than 0.1): Near Independence**

**Stations and Weather Conditions**

Minimal associations (V = 0.049-0.081) between stations and weather descriptors align with geographic and meteorological principles specific to Hanoi. Synoptic-scale weather systems (100-1000 km spatial extent) dwarf Hanoi's urban area (approximately 58 km diameter), with station spacing of 5-15 km. This scale disparity ensures all monitoring stations observe essentially uniform weather conditions across the city. The near-independence (less than 10 percent shared variance) confirms stations represent metadata rather than predictive features.

Small non-zero associations arise from technical factors rather than spatial weather variation: instrumental noise and calibration differences between stations, temporal reporting lags, and minor microclimate effects. Geographic context reinforces this interpretation - Hanoi's flat terrain and compact size minimize local weather heterogeneity. Consequently, stations can be safely excluded from predictive models without information loss.

**Feature Engineering Strategy**

**Sunrise and Sunset Transformation**

Categorical sunrise/sunset strings require conversion to numerically meaningful features capturing seasonal patterns. Recommended transformations include:

Day length (primary recommendation): Compute hours between sunset and sunrise, yielding interpretable values ranging from 11.5 hours (winter solstice) to 13.5 hours (summer solstice), directly capturing seasonal variation.

Solar timing variables: Extract sunrise and sunset as numerical hours (e.g., 6.75 for 6:45 AM), useful for solar angle calculations and energy modeling applications.

Cyclical encoding: Transform day of year into sine and cosine components (using 2 pi multiplied by day of year divided by 365.25), capturing seasonal periodicity while avoiding December 31 to January 1 discontinuity.

Solar position parameters (advanced): Calculate solar noon as temporal midpoint and solar declination angle for specialized applications including solar energy and agricultural modeling.

These transformations convert two problematic categorical variables (V = 1.0 spurious correlation) into one to four physically meaningful numerical features, eliminating multicollinearity while preserving seasonal information and enabling model generalization.

**Weather Descriptor Consolidation**

Two approaches address weather variable redundancy:

Option 1 (Retention): Keep conditions, drop description and icon, then apply one-hot encoding to create 10-15 binary features. This preserves categorical weather information with interpretable effects (Clear, Rain, Cloudy) but increases dimensionality and may duplicate information already captured by numerical variables (cloudcover, precipprob, precipcover).

Option 2 (Elimination - Recommended): Drop all three weather descriptors, relying instead on numerical equivalents. Rationale includes: conditions information adequately captured by continuous variables (Rain approximated by high precipprob and precipcover, Clear by low cloudcover), reduced dimensionality improving model efficiency, and enhanced robustness through consistent use of continuous measurements. This approach prioritizes simplicity and generalizability over categorical interpretability.

**Stations Removal**

Station identifiers warrant complete removal from predictive models given minimal weather associations (V less than 0.1), spurious sunrise/sunset correlations (V = 1.0), and pure metadata status. Hanoi's compact geography ensures spatial homogeneity across all monitoring locations. Retention is justified only for data provenance tracking, quality control analysis identifying problematic instruments, or future multi-city expansion requiring spatial heterogeneity modeling.

**Implementation Summary**

Post-transformation feature set includes: one to three derived numerical features from sunrise/sunset (day length, seasonal encodings), optional 10-15 binary features from conditions encoding (if retention approach selected), and complete removal of description, icon, and stations. This strategy eliminates spurious correlations, preserves physically meaningful seasonal patterns, and substantially improves model stability and generalization capacity.

**Data Quality Insights**

Spurious perfect correlations (V = 1.0) stem from data structure characteristics rather than collection errors: daily temporal granularity creating unique sunrise identifiers, categorical encoding of continuous variables, and date as hidden confounder. Weak station-weather associations (V less than 0.1) confirm Hanoi's geographic homogeneity, with synoptic-scale systems dominating local conditions and flat terrain minimizing microclimate variation. Perfect weather descriptor correlations (V approximately 1.0) reflect intentional API design generating multiple format representations from single source classification. These patterns inform feature engineering priorities rather than indicating data quality deficiencies.

### Correlation Ratio Analysis: Numerical-Categorical Relationships

**Overview**

The correlation ratio, or eta-squared, quantifies the proportion of variance in a numerical variable explained by a categorical variable. Values range from 0 (no explanatory power) to 1 (perfect grouping), calculated via one-way ANOVA comparing between-group to total variance.

**Key Findings**

**Strong Associations (eta-squared greater than or equal to 0.5)**

UV Index demonstrates exceptional explanatory power for solar radiation and energy variables (eta-squared approximately 0.93-0.94), reflecting their shared dependence on solar zenith angle, atmospheric optical depth, and cloud cover. This strong physical relationship suggests UV Index categories can serve as an effective proxy for solar radiation in simplified models while enhancing interpretability for non-technical stakeholders.

Severe Risk exhibits substantial associations with temperature extremes, with feels-like maximum temperature showing stronger correlation (eta-squared = 0.657) than actual temperature (eta-squared = 0.487). This 17% differential underscores the role of humidity and wind in heat stress and atmospheric instability. The relationship aligns with meteorological principles of Convective Available Potential Energy (CAPE), where high temperature combined with high moisture creates conditions conducive to severe weather events. Similarly, sea level pressure shows moderate-to-strong association with Severe Risk (eta-squared = 0.553), consistent with the linkage between low-pressure systems and convective activity.

Weather condition categories moderately explain cloud cover variance (eta-squared = 0.59-0.61), though substantial residual variance (38-41%) remains due to imprecise category boundaries and temporal aggregation. Standardized threshold-based classification would improve categorical homogeneity.

**Moderate Associations (0.3 to 0.5)**

Humidity exhibits weak-to-moderate relationships with weather descriptors (eta-squared = 0.32-0.38), reflecting Hanoi's subtropical monsoon climate characterized by persistently high baseline humidity (75-85%) regardless of weather conditions. This limited variability, influenced by proximity to the Red River and monsoon moisture advection, suggests dew point would provide superior predictive value for precipitation modeling.

Precipitation coverage shows moderate association with weather conditions (eta-squared = 0.544), though 46% unexplained variance stems from heterogeneity in precipitation intensity, spatial coverage, and temporal duration within categorical labels.

**Weak Associations (eta-squared less than 0.3)**

Temperature variables demonstrate poor associations with weather conditions (eta-squared approximately 0.16-0.21), attributable to dominant seasonality effects. Weather categories fail to account for seasonal context, where identical conditions correspond to vastly different temperatures across seasons. This necessitates interaction terms combining conditions with temporal features for effective temperature modeling.

Wind variables exhibit negligible associations with all categorical variables (eta-squared less than or equal to 0.06), confirming that Hanoi's low-wind regime (mean 2-4 meters per second) is primarily governed by synoptic-scale pressure gradients and monsoon circulation rather than local weather conditions. The high noise-to-signal ratio justifies excluding wind variables from predictive models unless specifically forecasting wind parameters.

Lunar phase shows no meaningful relationship with meteorological variables (eta-squared approximately 0.003), as expected from physical principles, and can be safely removed from feature sets.

**Spurious Correlations**

Sunrise and sunset exhibit perfect correlation (eta-squared = 1.0) with all numerical variables, representing a methodological artifact rather than causal relationship. This spurious association arises from temporal granularity, where daily aggregation creates perfect cardinality matching between observations and unique sunrise/sunset times. These categorical variables should be excluded from models to avoid perfect multicollinearity. Recommended alternatives include derived features with physical significance: day length (sunset minus sunrise), seasonal indices using sinusoidal encoding, and solar noon calculations.

**Feature Engineering Recommendations**

High-priority categorical predictors include UV Index (solar radiation proxy), Severe Risk (temperature and pressure anomalies), and weather conditions (cloud cover and precipitation coverage only). Low-value features warranting removal include sunrise/sunset categories (spurious correlation), moonphase (no signal), and wind variables (predominantly noise in Hanoi context).

For temperature and seasonally-dependent variables, interaction features combining weather conditions with temporal components (month, day of year sine/cosine transformations) are essential. Regression models should prioritize variables with eta-squared exceeding 0.5 relative to target variables and employ regularization techniques to mitigate effects of low-variance predictors.

**Limitations**

High eta-squared values do not imply causation, as demonstrated by the sunrise/sunset artifact. Substantial within-category variance (40-60%) across multiple relationships indicates potential benefits from refining categorical definitions to increase group homogeneity. Findings reflect Hanoi's subtropical monsoon climate and may not generalize to other climate regimes.
