## Correlation Analysis for Categorical and Mixed Variables

### 1. Categorical-Categorical Correlation: CramÃ©r's V
CramÃ©r's V Ä‘o lÆ°á»ng má»‘i quan há»‡ giá»¯a 2 biáº¿n categorical dá»±a trÃªn chi-square test. GiÃ¡ trá»‹ tá»« 0 (khÃ´ng cÃ³ má»‘i quan há»‡) Ä‘áº¿n 1 (má»‘i quan há»‡ hoÃ n háº£o).

### 2. Numerical-Categorical Correlation: Correlation Ratio (Î·Â²)
Correlation ratio (eta-squared) Ä‘o lÆ°á»ng má»©c Ä‘á»™ má»™t biáº¿n categorical giáº£i thÃ­ch Ä‘Æ°á»£c sá»± biáº¿n thiÃªn cá»§a má»™t biáº¿n numerical. GiÃ¡ trá»‹ tá»« 0 Ä‘áº¿n 1.

```python
# Import thÆ° viá»‡n cáº§n thiáº¿t
import numpy as np
from scipy.stats import chi2_contingency
import warnings
warnings.filterwarnings('ignore')

# HÃ m tÃ­nh CramÃ©r's V cho categorical-categorical correlation
def cramers_v(x, y):
    """
    TÃ­nh CramÃ©r's V giá»¯a 2 biáº¿n categorical
    V = sqrt(chi2 / (n * (min(k,r) - 1)))
    """
    # Loáº¡i bá» missing values
    mask = ~(pd.isna(x) | pd.isna(y))
    x_clean = x[mask]
    y_clean = y[mask]
    
    if len(x_clean) == 0:
        return np.nan
    
    # Táº¡o contingency table
    confusion_matrix = pd.crosstab(x_clean, y_clean)
    
    # Chi-square test
    chi2, _, _, _ = chi2_contingency(confusion_matrix)
    
    n = confusion_matrix.sum().sum()
    min_dim = min(confusion_matrix.shape[0], confusion_matrix.shape[1]) - 1
    
    if min_dim == 0:
        return 0
    
    v = np.sqrt(chi2 / (n * min_dim))
    return v

# HÃ m tÃ­nh correlation ratio (eta-squared) cho numerical-categorical correlation
def correlation_ratio(categories, values):
    """
    TÃ­nh correlation ratio (Î·Â²) giá»¯a biáº¿n categorical vÃ  numerical
    Î·Â² = SSB / SST (Sum of Squares Between / Sum of Squares Total)
    """
    # Loáº¡i bá» missing values
    mask = ~(pd.isna(categories) | pd.isna(values))
    categories_clean = categories[mask]
    values_clean = values[mask]
    
    if len(values_clean) == 0:
        return np.nan
    
    # TÃ­nh mean tá»•ng thá»ƒ
    mean_total = values_clean.mean()
    
    # TÃ­nh SST (Total Sum of Squares)
    sst = np.sum((values_clean - mean_total) ** 2)
    
    if sst == 0:
        return 0
    
    # TÃ­nh SSB (Between group Sum of Squares)
    ssb = 0
    for category in categories_clean.unique():
        group_values = values_clean[categories_clean == category]
        group_mean = group_values.mean()
        group_size = len(group_values)
        ssb += group_size * (group_mean - mean_total) ** 2
    
    # Î·Â² = SSB / SST
    eta_squared = ssb / sst
    return eta_squared

print("âœ… ÄÃ£ Ä‘á»‹nh nghÄ©a cÃ¡c hÃ m tÃ­nh correlation:")
```

### BÆ°á»›c 1: XÃ¡c Ä‘á»‹nh cÃ¡c loáº¡i biáº¿n (categorical vs numerical)

```python
# XÃ¡c Ä‘á»‹nh cÃ¡c biáº¿n categorical vÃ  numerical
categorical_cols = hn_daily.select_dtypes(include=['object', 'category']).columns.tolist()
# ThÃªm cÃ¡c biáº¿n cÃ³ sá»‘ lÆ°á»£ng unique nhá» (cÃ³ thá»ƒ coi nhÆ° categorical)
low_cardinality_threshold = 20
for col in hn_daily.columns:
    if col not in categorical_cols and hn_daily[col].nunique(dropna=True) <= low_cardinality_threshold:
        categorical_cols.append(col)

numerical_cols = hn_daily.select_dtypes(include=['float64', 'int64']).columns.tolist()
# Loáº¡i bá» cÃ¡c cá»™t low-cardinality khá»i numerical
numerical_cols = [col for col in numerical_cols if col not in categorical_cols]

print(f"ðŸ“Š Sá»‘ biáº¿n categorical: {len(categorical_cols)}")
print(f"Danh sÃ¡ch: {categorical_cols}\n")
print(f"ðŸ“ˆ Sá»‘ biáº¿n numerical: {len(numerical_cols)}")
print(f"Danh sÃ¡ch: {numerical_cols[:10]}...")  # In 10 biáº¿n Ä‘áº§u
```

### BÆ°á»›c 2: TÃ­nh Categorical-Categorical Correlation (CramÃ©r's V)

```python
# TÃ­nh CramÃ©r's V matrix cho táº¥t cáº£ cÃ¡c cáº·p biáº¿n categorical
cat_cat_corr = pd.DataFrame(
    np.zeros((len(categorical_cols), len(categorical_cols))),
    index=categorical_cols,
    columns=categorical_cols
)

print("ðŸ”„ Äang tÃ­nh CramÃ©r's V cho cÃ¡c cáº·p biáº¿n categorical...")
for i, col1 in enumerate(categorical_cols):
    for j, col2 in enumerate(categorical_cols):
        if i <= j:  # Chá»‰ tÃ­nh ná»­a trÃªn cá»§a ma tráº­n (symmetric)
            v = cramers_v(hn_daily[col1], hn_daily[col2])
            cat_cat_corr.loc[col1, col2] = v
            cat_cat_corr.loc[col2, col1] = v

print("âœ… HoÃ n thÃ nh!\n")
print("ðŸ“Š CramÃ©r's V Matrix (Categorical-Categorical):")
print(cat_cat_corr)
```

```python
# Visualize CramÃ©r's V matrix
if len(categorical_cols) > 0:
    plt.figure(figsize=(10, 8))
    sns.heatmap(cat_cat_corr, 
                annot=True, 
                fmt='.3f', 
                cmap='YlOrRd',
                square=True,
                linewidths=0.5,
                cbar_kws={"shrink": 0.8},
                vmin=0, vmax=1)
    plt.title("CramÃ©r's V - Categorical-Categorical Correlation", fontsize=14, fontweight='bold', pad=15)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
    
    # In ra cÃ¡c cáº·p cÃ³ correlation cao (V > 0.3)
    print("\nðŸ”¥ CÃ¡c cáº·p biáº¿n categorical cÃ³ CramÃ©r's V > 0.3:")
    high_corr = []
    for i in range(len(categorical_cols)):
        for j in range(i+1, len(categorical_cols)):
            v = cat_cat_corr.iloc[i, j]
            if v > 0.3:
                high_corr.append((categorical_cols[i], categorical_cols[j], v))
    
    if high_corr:
        for col1, col2, v in sorted(high_corr, key=lambda x: x[2], reverse=True):
            print(f"  â€¢ {col1} â†” {col2}: V = {v:.3f}")
    else:
        print("  KhÃ´ng cÃ³ cáº·p nÃ o cÃ³ V > 0.3")
```

### BÆ°á»›c 3: TÃ­nh Numerical-Categorical Correlation (Correlation Ratio Î·Â²)

```python
# TÃ­nh correlation ratio (Î·Â²) giá»¯a numerical vÃ  categorical
num_cat_corr = pd.DataFrame(
    np.zeros((len(numerical_cols), len(categorical_cols))),
    index=numerical_cols,
    columns=categorical_cols
)

print("ðŸ”„ Äang tÃ­nh Correlation Ratio (Î·Â²) cho cÃ¡c cáº·p numerical-categorical...")
for num_col in numerical_cols:
    for cat_col in categorical_cols:
        eta = correlation_ratio(hn_daily[cat_col], hn_daily[num_col])
        num_cat_corr.loc[num_col, cat_col] = eta

print("âœ… HoÃ n thÃ nh!\n")
print("ðŸ“Š Correlation Ratio Matrix (Numerical-Categorical):")
print(num_cat_corr.head(10))  # In 10 hÃ ng Ä‘áº§u
```

```python
# Visualize Numerical-Categorical correlation
if len(numerical_cols) > 0 and len(categorical_cols) > 0:
    # Giá»›i háº¡n sá»‘ hÃ ng hiá»ƒn thá»‹ Ä‘á»ƒ dá»… nhÃ¬n
    max_rows = min(20, len(numerical_cols))
    
    plt.figure(figsize=(12, max(8, max_rows * 0.4)))
    sns.heatmap(num_cat_corr.iloc[:max_rows], 
                annot=True, 
                fmt='.3f', 
                cmap='RdYlGn',
                linewidths=0.5,
                cbar_kws={"shrink": 0.8}, 
                vmin=0, vmax=1)
    plt.title(f"Correlation Ratio (Î·Â²) - Numerical-Categorical (Top {max_rows} vars)", 
              fontsize=14, fontweight='bold', pad=15)
    plt.xlabel('Categorical Variables')
    plt.ylabel('Numerical Variables')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
    
    # TÃ¬m cÃ¡c cáº·p cÃ³ correlation cao (Î·Â² > 0.3)
    print("\nðŸ”¥ CÃ¡c cáº·p biáº¿n numerical-categorical cÃ³ Î·Â² > 0.3:")
    high_corr = []
    for num_col in numerical_cols:
        for cat_col in categorical_cols:
            eta = num_cat_corr.loc[num_col, cat_col]
            if eta > 0.3:
                high_corr.append((num_col, cat_col, eta))
    
    if high_corr:
        for num_col, cat_col, eta in sorted(high_corr, key=lambda x: x[2], reverse=True)[:20]:
            print(f"  â€¢ {num_col} â†” {cat_col}: Î·Â² = {eta:.3f}")
    else:
        print("  KhÃ´ng cÃ³ cáº·p nÃ o cÃ³ Î·Â² > 0.3")
```
